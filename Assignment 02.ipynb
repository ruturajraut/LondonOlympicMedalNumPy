#### 1: View and add the dataset

```python
#Import the necessary library
import numpy as np
```

```python
#Manually add the Summer Olympics, London 2012 dataset as arrays
Countries = np.array(['GBR','China','RUS','US','KOR','JPN','GER'])
GoldMedal = np.array([29,38,24,46,13,7,11])
SilverMedal = np.array([17,28,25,28,8,14,11])
BronzeMedal = np.array([19,22,32,29,7,17,14])
```

#### Find the country with maximum gold medals

```python
#Use the argmax() method to find the highest number of gold medals
max_gold_index = GoldMedal.argmax()
max_gold_index
```

```python
#Print the name of the country
country_with_max_gold_medal = Countries[max_gold_index]
country_with_max_gold_medal
```

#### Find the countries with more than 20 gold medals

```python
#Use Boolean indexing technique to find the required output
print(Countries[GoldMedal>20])
```

#### Evaluate the dataset and print the name of each country with its gold medals and total number of medals

```python
#Use a for loop to create the required output
for i in range(len(Countries)):
    gold_medal = GoldMedal[i]
    country = Countries[i]
    total_medal = BronzeMedal[i]+GoldMedal[i]+SilverMedal[i]
    print('{}, gold medal {}, total medals {} '.format(country,gold_medal,total_medal))
```
